```python
import matplotlib.pyplot as plt
import random
from time import perf_counter
import numpy as np
```

```python
def list_summation(S):
    total = 0
    for i in range(len(S)):
        total+= S[i]
    return total
```

```python
def recursive_summation(S):
    length = len(S)
    if length == 0:
        return 0
    else:
        return S[length - 1] + recursive_summation(S[0:length - 1])
```

```python
def timeit(func, S):
    before = perf_counter()
    func(S)
    after = perf_counter()
    return after - before
```

```python
def list_generator(n):
    return [random.randint(0,1000) for i in range(0, n)]
```

```python
def time_counts_generator(func, set_generator, size):
    time_counts = []
    for i in range(100, size):
        time_counts.append(timeit(func, set_generator(size)))
    return time_counts
```

### Instantiating List of Time Counts for `recursive_summation`.

```python
time_counts = time_counts_generator(recursive_summation, list_generator, 1000)
```

```python
plt.title("Rate of Change of Time Taken for Execution")
plt.plot([i + 1 for i in range(len(time_counts))], time_counts)
```

### Instantiating List of Time Counts for `list_summation`.

```python
time_counts = np.array(time_counts_generator(list_summation, list_generator, 1000))
```

```python
plt.title("Rate of Change of Time Taken for Execution")
plt.plot([i + 1 for i in range(len(time_counts))], time_counts)
```
